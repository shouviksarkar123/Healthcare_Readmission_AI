# Data Governance – Healthcare Readmission AI

## Purpose
Ensure secure, controlled, and compliant access to sensitive healthcare data
while enabling trusted analytics and machine learning.

## Governance Strategy
- Data organized using Medallion Architecture (Bronze → Silver → Gold)
- Raw patient data restricted to engineering layers
- Business users access only aggregated, privacy-safe Gold views
- Role-based access concepts applied to prevent data leakage

## Business Value
- Protects patient privacy and reduces compliance risk
- Builds trust in analytics and AI predictions
- Enables safe decision-making without exposing sensitive data

```sql
-- Create a business-safe aggregated view
-- No patient-level or PII data exposed
-- Uses cleaned Gold layer column: readmit_30d

CREATE OR REPLACE VIEW gold_readmission_summary_view AS
SELECT
  readmit_30d,                  -- 1 = readmitted within 30 days, 0 = not readmitted
  COUNT(*) AS patient_count     -- total number of patients in each category
FROM gold_patient_features
GROUP BY readmit_30d;
```
